```python
import torch
```

张量代表一个数值组成的数组，这个数组有多个维度

```python
x = torch.arange(12)
x
```

可以通过shape属性访问张量的形状和元素的个数

```python
x.shape,x.numel()
```

可以通过reshape改变张量的形状，不改变里面的数据

```python
x = x.reshape(3,4)
x
```

使用全0、全1、其他常量或者从特定分布中随机采样的数字

```python
torch.zeros((2,3,4)), torch.ones((2,3,4))
```

可以用Python列表（或嵌套列表）来为所需的张量赋予确定的值

```python
torch.tensor([[2,1,4,3],[1,2,3,4],[4,3,2,1]])
```

可以进行算术运算符（+ - * / **）的计算，一一对应

```python
x = torch.tensor([1.0,2,4,8])
y = torch.tensor([2,2,2,2])
x+y, x-y, x*y, x/y,x**y
```

```python
#也有更多的指数运算
torch.exp(x)
```

可以把多个张量连接在一起

```python
x = torch.arange(12,dtype=torch.float32).reshape(3, 4)
y = torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((x,y),dim=0), torch.cat((x,y),dim=1)
```

通过逻辑运算符构建二元张量

```python
x == y
```

对张量中所有的元素求和会产生一个只有一个元素的张量

```python
x.sum()
```

形状不同，也可通过调用广播机制（broadcaing mechanism）来执行按元素操作

```python
a = torch.arange(3).reshape((3,1))
b = torch.arange(2).reshape((1,2))
#将第一个张量a复制成3x2的形式，将第二个张量b复制成3x2的形式，然后对应元素相加
a,b,a+b
```

可以使用下标来访问数据

```python
x[-1], x[1:3]
```

也可以使用下标进行写值

```python
x[1,2] = 9
x
```

```python
x[0:2, :] = 12
x
```

运行一些操作可能会重新分配内存

```python
before = id(y)
Y = y + x
id(Y) == before
```

原地操作

```python
z = torch.zeros_like(y)
print('id(z):', id(z))
z[:] = y + x
print('id(z):', id(z))
```

如果在后续计算中没有重复使用x，我们也可以使用x[:]=x+Y或x十=Y来减少操作的内存开销。

```python
before = id(x)
x += y 
id(x) == before
```
